```python
import numpy as np

k = 10

n_steps = 2000
epsilon = 0.1
rewards = np.zeros(n_steps) # array 
action_values = np.zeros(k)  # Estimated rewards for each arm
action_counts = np.zeros(k)     # Number of times each arm is selected

true_action_values = np.random.normal(0, 1, k)  # True mean rewards for each arm
```

```python
for step in range(n_steps):
    #select an action using the epsilon-greedy policy:
    if np.random.rand() < epsilon :
        action = np.random.choice(k)
    else:
        action = np.argmax(action_values)

    #compute the reward
    reward = np.random.normal(true_action_values[action],1)

    #update the estimated actiom-value function Q(a) using the sample-average method
    action_counts[action] += 1
    action_values[action] += (reward - action_values[action]) / action_counts[action]

    rewards[step]+= reward

```

```python
import matplotlib.pyplot as plt


# Plotting the rewards over time
plt.figure(figsize=(10, 6))
plt.plot(range(n_steps), rewards, label=f'Reward')
plt.xlabel('Steps')
plt.ylabel('Reward')
plt.title('Rewards Over Time for 10-Armed Bandit Problem')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Arrays for rewards, action values, and counts
rewards = np.zeros(n_steps)  # To store rewards at each step
action_values = np.zeros(k)  # Estimated rewards for each arm
action_counts = np.zeros(k)  # Number of times each arm is selected

# True action values (true mean rewards for each arm)
true_action_values = np.random.normal(0, 1, k)

# Find the optimal action (arm with the highest true action value)
optimal_action = np.argmax(true_action_values)

# To track optimal actions over time
optimal_action_count = np.zeros(n_steps)

# Simulation loop
for step in range(n_steps):
    # Select an action using the epsilon-greedy policy
    if np.random.rand() < epsilon:
        action = np.random.choice(k)  # Explore: random action
    else:
        action = np.argmax(action_values)  # Exploit: choose the best estimated action

    # Compute the reward
    reward = np.random.normal(true_action_values[action], 1)

    # Update the estimated action-value function Q(a) using the sample-average method
    action_counts[action] += 1
    action_values[action] += (reward - action_values[action]) / action_counts[action]

    # Store the reward for this step
    rewards[step] = reward

    # Track if the optimal action was chosen
    if action == optimal_action:
        optimal_action_count[step] = 1

# Calculate the percentage of optimal actions over time
optimal_action_percentage = np.cumsum(optimal_action_count) / (np.arange(n_steps) + 1)

# Plot the percentage of optimal actions over time
plt.plot(optimal_action_percentage)
plt.xlabel('Steps')
plt.ylabel('Optimal Action Percentage')
plt.title('Optimal Action Percentage over Time')
plt.show()
```
